```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
```

**line**

```
a=[2,2,8,5,4]
b=[5,3,7,9,3]
plt.plot(a,b)
plt.show()
```

**tagging with label**

```
a=[2,2,8,5,4]
b=[5,3,7,9,3]
plt.plot(a,b,linewidth=5)
plt.xlabel('Food')
plt.ylabel('Yummy')
plt.title('Taste analysis')
plt.show()
```

**Legend in graph**

```
a=[2,2,8,5,4]
b=[5,3,7,9,3]
plt.plot([2,6,4],[7,5,6],label='May')
plt.plot(a,b,label='June')
plt.xlabel('Food')
plt.ylabel('Yummy')
plt.title('Report of JUNE')
plt.legend()
plt.show()
```

**Plotting bar chart**

```
plt.bar([2,6,4,7,9],[9,4,3,6,4],label='june month',color='y')
plt.bar([3,5,2,4,6],[4,5,8,2,3],label='may month', color='r')
plt.xlabel('Food')
plt.ylabel('Yummy')
plt.title('taste analusis as bar chart')
plt.legend()
plt.grid(True,color='g')
plt.show()
```

```
a1=np.arange(0,100,2.5)
a2=[i for i in a1]
plt.bar(a1,a2)
plt.grid(True,color='g')
```

```
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

```
men_means, men_std = (20, 35, 30, 35, 27), (2, 3, 4, 1, 2)
women_means, women_std = (25, 32, 34, 20, 25), (3, 5, 2, 3, 3)

ind = np.arange(len(men_means))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, men_means, width, yerr=men_std,
                label='Men')
rects2 = ax.bar(ind + width/2, women_means, width, yerr=women_std,
                label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()
```

```
data = {'apples': 10, 'oranges': 15, 'lemons': 5, 'limes': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Categorical Plotting')
```

**plotting histogram**

```
x=[3,7,1,8,3,9,10,43]
y=[2,66,22,8,1,99,3,55]
plt.title("simple")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.scatter(x,y,label="hello",color='b',marker='.',s=200)
# marker can be - *,o,x, 
#  only x , y in scattre is important 
#  when both parameter have same number of element then we use scatter
plt.legend()
plt.grid(True,color='y')
plt.show()
```

**multiple lines**

```
x=np.arange(1,20,0.5)
plt.plot(x,x,'ro',x,x**2,'bs',x,x**3,'gv',x,x**4,'y--')
plt.show()
```

**Pie chart**

```
import  matplotlib.pyplot  as plt
import numpy as np


slices=[3,5,9,4]
activities=['sleep','eat','code','play']
colors=['b','y','g','m']

plt.pie(slices,labels=activities,colors=colors,startangle=90,
        explode=(0.1,0.2,0,0),autopct='%1.1f%%',shadow=True)
# startangle is optional 
# shadow=True is also optinal
# explode=(0,0.1,0,0)
plt.title("simple")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.legend()
plt.grid(True,color='y')
plt.show()
```
