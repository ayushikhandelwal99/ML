```
import matplotlib.pyplot as plt
```

```
x=[2,3,7,3]
x1=[9,5,3,6]
y1=[7,1,2,3]
y=[9,5,4,0]
plt.xlabel("time")  #this will label x-axis as time
plt.ylabel("speed") #this will label y-axis as speed
plt.plot(x,y,'bs--',label="sand") #this will draw the graph with blue colour,square points and dashed line with label sand
plt.plot(x1,y1,'ro--',label="water")
plt.grid(color='green') #change the grid colour to green
plt.legend() #this is used to show labels
plt.xlim(0,12)
plt.ylim(0,15) # this will set the limit of values
plt.bar(x,y)
plt.bar(x1,y1)
```

```
x=[2,3,7,3]
x1=[9,5,3,6]
y1=[7,1,2,3]
y=[9,5,4,0]
plt.xlabel("time")  #this will label x-axis as time
plt.ylabel("speed") #this will label y-axis as speed
plt.plot(x,y,label="sand") #this will draw the graph with blue colour,square points and dashed line with label sand
plt.plot(x1,y1,label="water")
plt.grid(color='green') #change the grid colour to green
plt.legend() #this is used to show labels
plt.xlim(0,12)
plt.ylim(0,15) # this will set the limit of values
plt.bar(x,y)
plt.bar(x1,y1)
```
