```
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
import sys
```

```
abc=cv2.imread('dog.jpg',1)  #1 means img in same colour --GRB or RGB
abc1=cv2.imread('dog.jpg',0)#0 means nocolour channel i.e. black and white -- gray image
abc2=cv2.imread('dog.jpg',-1)#-1 maintains transparency
```

```
# Get number of pixel horizontally and vertically. 
    (height, width) = abc.shape[:2] 
  
    # Specify the size of image along with interploation methods. 
    # cv2.INTER_AREA is used for shrinking, whereas cv2.INTER_CUBIC 
    # is used for zooming. 
    res = cv2.resize(abc, (int(width / 2), int(height / 2)), interpolation = cv2.INTER_CUBIC)
    cv2.imwrite('result.jpg',res)
```
